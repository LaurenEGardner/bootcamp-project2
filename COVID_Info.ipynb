```python
import numpy as np
import pandas as pd
import datetime as dt
```

```python
dataFile = "Resources/covid_us_states.csv"
```

```python
covid_df = pd.read_csv(dataFile)
covid_df.head()
```

```python
covid_df.dtypes
```

```python
covid_df['day'] = pd.to_datetime(covid_df['day'])
covid_df.head()
```

```python
covid_df['year'] = covid_df['day'].dt.year
covid_df['month'] = covid_df['day'].dt.month
covid_df.head()
```

```python
december = covid_df[covid_df['month'] == 12]
december
```

```python
monthlyTotals = covid_df.loc[(covid_df['day'] == '2020-01-31') | (covid_df['day'] == '2020-02-29') | (covid_df['day'] == '2020-03-31') | (covid_df['day'] == '2020-04-30') | (covid_df['day'] == '2020-05-31') | (covid_df['day'] == '2020-06-30') | (covid_df['day'] == '2020-07-31') | (covid_df['day'] == '2020-08-31') | (covid_df['day'] == '2020-09-30') | (covid_df['day'] == '2020-10-31') | (covid_df['day'] == '2020-11-30') | (covid_df['day'] == '2020-12-31')]
monthlyTotals.head()
```

```python
usMonthlyTotals = monthlyTotals[['cases', 'deaths', 'month']]
usMonthlyTotals = usMonthlyTotals.groupby(['month']).sum()
usMonthlyTotals
```

```python
# Export to CSV
monthlyTotals.to_csv("Resources/covidData_clean.csv", encoding="utf-8", index=False)
usMonthlyTotals.to_csv("Resources/usCovidData.csv", encoding="utf-8", index=True)
```
