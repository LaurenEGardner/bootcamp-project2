```python
#Setting variables 
year='2020'
month='*'
cols='SUMMARY_LVL,SUMMARY_LVL2,CTY_CODE,CTY_NAME,CTY_SUBCODE,GEN_VAL_MO,GEN_VAL_YR,DISTRICT,DIST_NAME,CON_VAL_MO,CON_VAL_YR,I_COMMODITY_SDESC'
```

```python
#Base URL 
base_url = 'https://api.census.gov/data/timeseries/intltrade/imports/hs'

complete_url = f'{base_url}?get={cols}&YEAR={year}&MONTH={month}'


hts_list=['3926201010','4015110110']
df_list=[]

for each_hts in hts_list:
    hts_code=each_hts
    query_params={
        'get': 'SUMMARY_LVL,SUMMARY_LVL2,CTY_CODE,CTY_NAME,CTY_SUBCODE,GEN_VAL_MO,GEN_VAL_YR,DISTRICT,DIST_NAME,CON_VAL_MO,CON_VAL_YR,I_COMMODITY_SDESC',
        'I_COMMODITY': hts_code,
        'YEAR': '2020',
        'MONTH': "*"
    }
    import_response=requests.get(base_url, params=query_params)
    #print(max_response.json())
    df_list.append(import_response.json()[1:])
    #print(df_list)
```

```python
importList = []
for product in df_list:
    importList += product
```

```python
importData2020 = pd.DataFrame(importList, columns=import_response.json()[0])
importData2020
```

```python
# drop aggregated totals
importData2020= importData2020[importData2020['CTY_NAME']!="TOTAL FOR ALL COUNTRIES"]
importData2020= importData2020[importData2020['DIST_NAME']!="TOTAL FOR ALL DISTRICTS"]
importData2020
```

```python
importData2020['I_COMMODITY_SDESC'].unique()
```

```python
importData2020 = importData2020.rename(columns={                                            
    "SUMMARY_LVL": "SUMMARY_LEVEL", "SUMMARY_LVL2": "SUMMARY_LEVEL_2", 
    "CTY_CODE": "COUNTRY_CODE", "CTY_NAME": "COUNTRY_NAME", "CTY_SUBCODE": "COUNTRY_SUBCODE",
    "GEN_VAL_MO": "GENERAL_VALUES_MONTH", "GEN_VAL_YR": "GENERAL_VALUES_YEAR",
    "DIST_NAME": "DISTRICT_NAME", "CON_VAL_MO": "MONTHLY_CONSUMPTION_VALUE", 
    "CON_VAL_YR": "YEARLY_CONSUMPTION_VALUE", "I_COMMODITY_SDESC": "COMMODITY_DESCRIPTION", 
    "I_COMMODITY": "COMMODITY"})
importData2020
```

```python
# Save as a csv
# Using encoding="utf-8" to avoid issues later
importData2020.to_csv("Resources/importData2020.csv", encoding="utf-8", index=False)
```
